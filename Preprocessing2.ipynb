```python
data_obj__ = data_obj[['معدلك في الثانوية العامة',
 'التخصص في الثانوية العامة',
 'نوع المدرسة',
 'الوضع الاقتصادي',
 'طبيعة عمل ولي الامر',
 'طبيعة السكن',
 'هل كنت تمتلك جهاز حاسوب',
 'هل يمتلك ولي أمرك وسيلة نقل خاصة',
 'التحصيل العلمي للوالد',
 'التحصيل العلمي للوالدة',
 'نوع الجامعة',
 'كليتك؟',]]
```

```python
obj = data_obj__.drop('كليتك؟',axis=1)
```

```python
obj
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder(categories = [['60-69', '70-79', '80-89', '90-99'], 
                                              list(data_obj__['التخصص في الثانوية العامة'].unique()),
                                              ['حكومي','خاصة'],
                                              ['ضعيف', 'جيد', 'جيد جدا', 'ممتاز'],
                                              ['أعمال حرة', 'موظف حكومي', 'موظف خاص'],
                                              ['مُلك خاص', 'إيجار'],
                                              ['نعم', 'لا'],
                                              ['نعم', 'لا'],
                                              ['يقرأ ويكتب', 'اعدادية', 'بكالوريوس', 'ماجستير', 'دكتوراة'],
                                              ['تقرأ وتكتب', 'اعدادية', 'بكالوريوس', 'ماجستير', 'دكتوراة'],
                                              ['خاصة', 'حكومية']])
cat_encoded = ordinal_encoder.fit_transform(obj)
data_obj_ord = pd.DataFrame(cat_encoded,columns=list(data_obj__.columns)[:-1])
```

```python
data_obj_ord
```

```python
data_num
```

```python
lable = data_obj__[['كليتك؟']]
```

```python
data_processing = pd.concat([data_obj_ord,data_num,lable],axis=1)
data_processing.sample()
```

```python
data_processing.to_csv('datanotebook2.csv',index=False)
```
