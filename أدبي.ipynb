```python
import numpy as np
import pandas as pd
import seaborn as ses
import matplotlib.pyplot as plt

from sklearn import preprocessing

from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('data/أدبي.csv')
```

```python
data
```

```python
list(data['كليتك؟'].unique())
```

```python
le = preprocessing.LabelEncoder()
le.fit(list(data['كليتك؟'].unique()))

list(le.classes_)

lable = list(le.transform(list(data['كليتك؟'])))

list(le.inverse_transform([2, 2, 1]))
```

```python
data['lable'] = lable
```

```python
data.drop('كليتك؟',axis=1,inplace=True)
```

```python
data
```

```python
X = data.drop('lable',axis=1)
y = data['lable']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.30,random_state =123)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)
```

```python
models[['Accuracy']]
```

```python
models.to_csv('أدبي.csv')
```
