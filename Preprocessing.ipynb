```python
x = data['هل تعاني من مشاكل صحية في']
```

```python
look, hearing, physical_problems, mentality, No = [], [], [], [], []
for i in x:
    if 'النظر' in i:
        look.append(1)
    else:
        look.append(0)
    
    if 'السمع' in i:
        hearing.append(1)
    else:
        hearing.append(0)
    
    if 'مشاكل جسدية' in i:
        physical_problems.append(1)
    else:
        physical_problems.append(0)
    
    if 'عقلية' in i:
        mentality.append(1)
    else:
        mentality.append(0)
    
    if 'لا' in i:
        No.append(1)
    else:
        No.append(0)
```

```python
print(len(look))
print(len(hearing))
print(len(physical_problems))
print(len(mentality))
print(len(No))
```

```python
sum(mentality)
```

```python
data['مشاكل في النظر'] = look
data['مشاكل في السمع'] = hearing
data['مشاكل جسدية'] = physical_problems
data['لا يوجد مشاكل'] = No

data.drop('هل تعاني من مشاكل صحية في',axis=1, inplace=True)
```

```python
data
```

```python
x = data[['ما هوايتك']]
```

```python
x.replace(np.NaN, 'لا', inplace=True)
```

```python
x = x['ما هوايتك']

sport, reading, writing, analysis, solving, tech, no = [], [], [], [], [], [], []
for i in x:
    if 'لا' in i:
        no.append(1)
    else:
        no.append(0)
    if 'الرياضة' in i:
        sport.append(1)
    else:
        sport.append(0)
    
    if 'القراءة' in i:
        reading.append(1)
    else:
        reading.append(0)
    
    if 'الكتابة' in i:
        writing.append(1)
    else:
        writing.append(0)
    
    if 'التحليل' in i:
        analysis.append(1)
    else:
        analysis.append(0)
    
    if 'حل المسائل' in i:
        solving.append(1)
    else:
        solving.append(0)
        
    if 'التكنولوجيا' in i:
        tech.append(1)
    else:
        tech.append(0)
```

```python
print(len(sport))
print(len(reading))
print(len(writing))
print(len(analysis))
print(len(solving))
print(len(tech))
print(len(no))
```

```python
data['الرياضة هوايتي'] = sport
data['القراءة هوايتي'] = reading
data['الكتابة هوايتي'] = writing
data['التحليل هوايتي'] = analysis
data['حل المسائل هوايتي'] = solving
data['التكنولوجيا هوايتي'] = tech
data['لا يوجد هواية'] = no

data.drop('ما هوايتك',axis=1, inplace=True)
```
